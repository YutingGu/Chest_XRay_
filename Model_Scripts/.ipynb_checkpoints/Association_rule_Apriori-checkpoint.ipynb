```python
import warnings
warnings.filterwarnings("ignore",'DeprecationWarning')

import pandas as pd
import numpy as np
from six import StringIO
pd.set_option('display.max_columns', None)

from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

import seaborn as sns
import matplotlib.pylab as plt
```

```python
train_label = pd.read_csv('../Dataset/train_label.csv', index_col=False).drop(['FileName'],axis=1)
val_label = pd.read_csv('../Dataset/val_label.csv', index_col=False).drop(['FileName'],axis=1)
test_label = pd.read_csv('../Dataset/test_label.csv', index_col=False).drop(['FileName'],axis=1)
```

```python
label_data = {'train': train_label, 'validation': val_label, 'test': test_label}
```

```python
def plot_apriori_heatmap(label_data_list):
    frames = []
    title = ''
    for t in label_data_list:
        frames.append(label_data[t])
        title += t + '|'

    all_label = pd.concat(frames)

    print('dataframe shape: ', all_label.shape)
    number_of_diagnosis = all_label.sum(axis=1)
    multilabel_idx = number_of_diagnosis[number_of_diagnosis>1].index
    all_label_subset = all_label.iloc[multilabel_idx]
    print('multilabel_subset shape: ', all_label_subset.shape)
    #train_label_sampling = train_label.sample(n=5000, random_state=0)

    # min_support = 0.1 means each diagnosis appear in at least 10% of patients
    frequent_labelsets = apriori(all_label_subset, min_support = 0.01, use_colnames=True)
    rules = association_rules(frequent_labelsets, metric='lift', min_threshold=1) 
    # rules.head(20)

    support_A = pd.DataFrame(columns=all_label_subset.columns, index=all_label_subset.columns)

    for index, row in rules.iterrows():
        if len(row['antecedents']) == 1 and len(row['consequents']) == 1:
            label_1, = row['antecedents']
            label_2, = row['consequents']
            support_A[label_1][label_2] = row['antecedent support']

    support_A = support_A.fillna(0)

    ax = sns.heatmap(support_A, linewidth=0.5, cmap="YlGnBu")
    ax.set_title(title)
    plt.show()
```

```python
all_combination = [['train'],['validation'],['test'],
                  ['train','validation'],['train','test'],['validation','test'],
                  ['train','validation','test']]
```

```python
for a in all_combination:
    plot_apriori_heatmap(a)
```
