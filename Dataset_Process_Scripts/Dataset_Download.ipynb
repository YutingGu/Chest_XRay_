```python
# Download the compressed image files
for idx, link in enumerate(links):
    filename = '../Dataset/images_%02d.tar.gz' % (idx+1)
    print('downloading '+filename+'...')
    urllib.request.urlretrieve(link, filename)  # download the zip file

print("Download complete. Please check the checksums")
```

```python
# Extracting all images from the compressed files
for idx, link in enumerate(links[1:]):
    filename = '../Dataset/images_%02d.tar.gz' % (idx+2)
    print('extracting '+filename+'...')
    tf = tarfile.open(filename, 'r')
    tf.extractall('Dataset')

print("Extracting complete. All images are now under Dataset/images folder.")
```

```python
# Show a random image
%matplotlib inline
import numpy as np                     
import matplotlib.pyplot as plt
import os
import cv2 as cv
import random
```

```python
def load_image(path):
    for img in os.listdir(path):
        print('Image name =',img)
        image = cv.imread(os.path.join(path, img))
        break
        
    return image
```

```python
im_path = '../Dataset/images/test_no_label'
img = load_image(im_path)
```
